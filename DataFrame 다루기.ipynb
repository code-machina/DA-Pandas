### .astype 메서드에서 대해 알아보자


````python
# dropna 메서드는 NaN 이 포함된 데이터를 버린다.
nba = pd.read_csv('nba.csv').dropna(how = 'all')
# College, Salary 역시 NaN 값을 0 과 Nan 으로 각각 치환한다.
nba['Salary'].fillna(0, inplace= True)
nba['College'].fillna("None", inplace=True)
nba.head(6)
````

````python
# 데이터 타입을 변환하기 위해 dtypes 어트리뷰트를 통해 데이터 타입을 확인한다.
# 이전 강의에서 살펴본 것처럼 nba.info() 메서드는 각 칼럼에 대한 요약 정보를 준다. 
# 가령 Name 칼럼은 null 이 아닌 데이터가 457 개가 있음을 의미한다. 그리고 이 데이터 타입은 object 이다.
nba.dtypes
nba.info()
````

````python
# astype 메서드는 inplace 옵션이 없으므로 재할당 해야한다.
nba['Salary'] = nba['Salary'].astype("int")
````

````python
# 다시 요약 정보를 보면 Salary 가 float 에서 int 로 변한 것을 확인할 수 있다. 
nba.info()
````

````python
nba['Number'] = nba['Number'].astype('int')
nba['Age'] = nba['Age'].astype('int')
nba.head(3)
````

````python
# .nunique() 메서드를 통해 특정 칼럼의 고유한 데이터가 몇개 있는지 확인 가능하다.
nba['Position'].nunique()
````

````python
# 참고로 category 타입으로 변환할 경우 데이터 크기가 줄어든다. 
nba['Position'] = nba['Position'].astype('category')
````

````python
# 그러나 변환한 결과에는 큰 차이가 없다.
nba.head(4)
````

````python
# NBA 에는 30 개의 팀이 있다.
nba['Team'].nunique()
nba['Team'] = nba['Team'].astype('category')
# 또 다시 데이터 크기가 줄었다. 좋은 처리 작업이다. 
nba.info()
````

### .sort_values 메서드를 사용하여 DataFrame 을 정렬하기



````python
nba = pd.read_csv('nba.csv')
nba.head(3)
````

````python
# Name 칼럼을 알파벳 순으로 정렬한다.
nba.sort_values('Name', ascending=False)
nba.sort_values('Age', ascending=False)
nba.sort_values('Salary', ascending=False, inplace=True) # Kobe Bryant 의 연봉이 제일 높다! 
nba.head(3)
````

````python
# Salary 데이터에 NaN 데이터가 많이 있다. 이 데이터는 na_position = ['first', 'last'] 등으로 지정할 수 있다.
# 그리고 여기에 ascending = False 를 추가할 경우 nan 데이터 그룹과 그렇지 않은 그룹을 두개로 나누어 각각 
# 정렬하는 모습을 볼 수 있다.

nba.sort_values('Salary', na_position = 'first').head(3)
````

````python
# by 파라미터에 칼럼 이름 리스트를 전달할 수있으며 ascending 파라미터에 이와 일대일로 대응되도록 
# Boolean 리스트를 전달할 수 있다.
# Team 은 ascending False (내림차순) 이며, Name 은 ascending True (오름차순) 이 된다.
nba.sort_values(['Team', 'Name'], kind='quicksort', ascending= [False, True], inplace=True)
````

### .rank 메서드를 이용한 Rank 매기기

````python
nba = pd.read_csv('nba.csv').dropna(how = 'all')
nba['Salary'] = nba['Salary'].fillna(0).astype("int")
nba.head(3)
````

````python
# Sslary 를 기준으로 랭크를 매기고 이를 Salary Rank 칼럼에 새로 저장한다. 이 때 type 을 int 로 한다.
nba["Salary Rank"] = nba["Salary"].rank(ascending=False).astype("int")
````

````python
nba.head()
````

````python
nba.sort_values('Salary Rank').head(3)
````
